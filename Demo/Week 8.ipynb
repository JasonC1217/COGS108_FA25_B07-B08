#### What is TF-IDF?

**TF-IDF** stands for **Term Frequency–Inverse Document Frequency**. It is a numerical statistic that reflects how important a word is to a document in a collection (or corpus).

TF-IDF is commonly used in **text mining**, **information retrieval**, and **natural language processing** to convert textual data into meaningful numerical features.

#### Intuition

- **Term Frequency (TF)** measures how often a term appears in a document.
- **Inverse Document Frequency (IDF)** measures how unique or rare a term is across all documents.
- The idea is to **increase the weight** of terms that occur frequently in a document **but not in many other documents** (i.e., not too common).

#### Formula

Let:
- $ t $: a term (word)
- $ d $: a single document
- $ D $: the full corpus (set of documents)

#### Term Frequency (TF):
$
\text{TF}(t, d) = \frac{\text{Number of times } t \text{ appears in } d}{\text{Total number of terms in } d}
$

#### Inverse Document Frequency (IDF):
$
\text{IDF}(t, D) = \log\left(\frac{N}{1 + n_t}\right)
$
- $ N $: total number of documents in the corpus
- $ n_t $: number of documents where the term \( t \) appears
- (The `+1` avoids division by zero)

#### TF-IDF Score:
$
\text{TF-IDF}(t, d, D) = \text{TF}(t, d) \times \text{IDF}(t, D)
$

#### Why Use TF-IDF?

- Helps filter out **common words** (e.g., "the", "and", "is") that are less meaningful for distinguishing documents.
- Highlights **discriminative terms** that are more relevant or unique to individual documents.

---

#### Example Use Case

If the word *"freedom"* appears frequently in one president’s inaugural address but rarely in others, TF-IDF will assign it a **higher score** for that speech—indicating its importance in that context.

To get started on your TF-IDF analysis, you'll first want to **create a `TfidfVectorizer` object to transform your text data into vectors. Assign this `TfidfVectorizer` object to `tfidf`.**

In this object, you'll need to **pass five arguments to initialize a `TfidfVectorizer`**: 
* set to apply TF scaling: `sublinear_tf=True`
* analyze at the word-level: `analyzer='word'`
* set maximum number of unique words: ` max_features=2000`
* specify that you want to tokenize the data using the word_tokenizer from NLTK: `tokenizer=word_tokenize`
* remove English language stop words: `stop_words=stopwords.words("english")`

Note: Stop words are a set of commonly used words in a language,which are commonly used in Text Mining and Natural Language Processing (NLP) to eliminate words that are so widely used that they carry very little useful information.

````python
tfidf = TfidfVectorizer(sublinear_tf=True,
                        analyzer='word',
                        max_features=2000,
                        tokenizer=word_tokenize,
                        stop_words=stopwords.words("english"))
````

The `TfidfVectorizer` will transform raw text documents into a matrix where each row represents a document and each column represents one of the top 2000 words. The value in each cell is the TF-IDF score, which is a statistical measure used to evaluate the importance of a word to a document in a collection or corpus.

Now, it's time to calculate TF-IDF for words across our corpus of Inaugural addresses! 

To do this:

1. Use the `tfidf.fit_transform` function to calculate TF-IDF on your `text` variable. Use `.toarray()` to transform the output to a numpy array named `tfidf_array`. 
2. Generate a DataFrame `inaug_tfidf` by passing `tfidf_array` into `pd.DataFrame`.
2. Be sure that the columns are named with the columns of the words the values represent and your index in `inaug_tfidf` is the year of the address. The `get_feature_names` method from `tfidf` may help you accomplish the columns name assignment. And the `years` you created earlier may help you with the indices.

````python
# Transform the preprocessed text data into a TF-IDF matrix and convert it to a pandas DataFrame.
# 1. tfidf.fit_transform(text): This line applies the previously initialized TfidfVectorizer to the text data.
#    - 'fit_transform' function computes the TF-IDF scores for each word in the text data.
#    - The result is a sparse matrix where each row corresponds to a document and each column represents 
# a term from the TF-IDF vocabulary.

inaug_tfidf = pd.DataFrame(tfidf.fit_transform(text).toarray())

# 2. Convert the sparse matrix to a dense matrix using '.toarray()', and then into a DataFrame.
#    - The DataFrame 'inaug_tfidf' represents the TF-IDF features of the text data, where each row is 
# a document and each column is a TF-IDF feature.

inaug_tfidf.columns = tfidf.get_feature_names_out()

# 3. Assign column names to the DataFrame using 'tfidf.get_feature_names_out()'.
#    - Each column in the DataFrame is named after a term in the TF-IDF vocabulary.

inaug_tfidf.index = years

# 4. Set the DataFrame index using 'years'.
#    - This implies that 'years' is an iterable (probably a list or array) that contains the years 
# associated with each text document.
#    - Each row in the DataFrame is now associated with a specific year, making it easier to reference 
# the documents chronologically.
````

# Model Selection with Diabetes Dataset

First, let's load our data:

````python
data = load_diabetes(as_frame=True, scaled=False)
diabetes = data.data
diabetes['target'] = data.target
diabetes.head()
````

Our dataset contains a variety of health-related measurements collected from patients. These variables include basic demographics as well as several biochemical indicators measured through blood tests.

The features are:

- **`age`** — patient’s age in years  
- **`sex`** — biological sex of the patient  
- **`bmi`** — body mass index, a measure of body fat based on height and weight  
- **`bp`** — average blood pressure  
- **`s1`** — total serum cholesterol  
- **`s2`** — LDL (low-density lipoproteins)  
- **`s3`** — HDL (high-density lipoproteins)  
- **`s4`** — TCH (total cholesterol)  
- **`s5`** — LTG (log of serum triglycerides)  
- **`s6`** — GLU (blood glucose level)  

Our **target variable** is:

- **`target`** — result of the **Oral Glucose Tolerance Test (OGTT)**, which reflects the progression of diabetes.

### What is OGTT?

An OGTT is a clinical test where:

1. The patient fasts for several hours  
2. The patient consumes a fixed amount of glucose  
3. After 2 hours, their blood glucose level is measured  

The resulting value (in mg/dL) becomes the **`target`** we want to predict from the other features.

### Next Step: Data Transformations

Before we can train a machine learning model, we will perform a few transformations on our dataset. These preprocessing steps will help us prepare the data, improve model performance, and ensure that our numerical and categorical features are handled appropriately.

#### Transform `target` into a Categorical Variable

Doctors use OGTT (Oral Glucose Tolerance Test) values to classify a patient’s diabetes status. The OGTT result is currently stored as a **numerical** value, but for our classification task, we want to convert it into **categorical labels** based on medical guidelines.

Below are the standard OGTT ranges and their clinical classifications:

- **Normal:**  
  `target` < **140 mg/dL**

- **Prediabetes:**  
  **140–199 mg/dL**

- **Diabetes:**  
  `target` ≥ **200 mg/dL**

### Your Task

Using the rules above, convert the numerical `target` values into a new categorical column called **`c_target`** in the dataframe.  
Each patient should receive one of the following labels:

- `"normal"`
- `"prediabetes"`
- `"diabetes"`

This new categorical feature will serve as the label for our classification model.

When you're done, your dataframe will include both:

- the original numerical `target`
- the new categorical `c_target`

````python
def transform_target(value):
    if value >= 140 and value < 199:
        return 'prediabetes'
    elif value >= 200:
        return 'diabetes'
    else:
        return 'normal'

diabetes['c_target'] = diabetes['target'].apply(transform_target)
````

#### Transform `age` to categorical
Let's transform the numerical feature `age` to a categorical one. Split the `age` values into groups of 20 years:
`0-19 years`, `20-39 years`, `40-59 years`, `60+ years`. Save this transformation as a new column called `c_age` in the dataframe.

````python
def transform_age(value):
    if value < 20:
        return '0-19 years'
    elif value >= 20 and value < 40:
        return '20-39 years'
    elif value >= 40 and value < 60:
        return '40-59 years'
    else:
        return '60+ years'

diabetes['c_age'] = diabetes['age'].apply(transform_age)
````

#### Create Train and Test Splits

Now that we transformed `target`, let's split the dataset into train and test splits. First, split the columns in the dataframe into variables `X` and `y`. Remember, we want to predict `target` based on all the other columns in our dataframe. 

````python
X = diabetes.drop(columns=['target', 'c_target', 'age'])
y = diabetes['c_target']
````

````python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=42, stratify = y)
````

#### Creating a Pipeline

In this step, we’ll build a **machine learning pipeline** that puts all of our preprocessing and modeling steps together into one object. This makes our workflow cleaner, less error-prone, and easier to reuse.

Our pipeline will do the following:

1. **Apply standard scaling** to the numerical features  
2. **One-hot encode** the categorical features (like `c_target`) into binary vectors  
   - Each category becomes its own column  
   - Because categories are mutually exclusive, exactly one column is `1` and the others are `0` for each row  
3. **Train a K-Nearest Neighbors (KNN) classifier** using the transformed features  

You should:

- **Create 2 column selectors** using `selector`:
  - One to collect only **numerical features**
  - One to collect only **categorical features**
  - Then, use these selectors on `X` to get `numerical_columns` and `categorical_columns`.

- **Build a `preprocessor`** using `ColumnTransformer`:
  - Apply a **`StandardScaler`** (z-score normalization) to the numerical columns  
  - Apply a **`OneHotEncoder`** to the categorical columns  

- **Create a pipeline** using `Pipeline()` from `sklearn.pipeline`:
  - First step: `"preprocessing"` → your `preprocessor`  
  - Second step: `"classifier"` → a `KNeighborsClassifier()` object  

- **Instantiate the classifier** using:
  ```python
  KNeighborsClassifier()

````python
# Selectors for splitting columns by data type
# - numerical_columns_selector: selects all non-object (numeric) columns
# - categorical_columns_selector: selects all object-type (categorical) columns
numerical_columns_selector = selector(dtype_exclude=object)
categorical_columns_selector = selector(dtype_include=object)

# Apply selectors to X to get the actual column names
numerical_columns = numerical_columns_selector(X)
categorical_columns = categorical_columns_selector(X)

# Define preprocessing steps:
# - OneHotEncoder: converts categorical variables into 0/1 indicator columns
# - StandardScaler: applies z-score normalization to numerical columns
onehot = OneHotEncoder()
scaler = StandardScaler()

# ColumnTransformer applies different transformations to different column subsets
# - onehot applied to categorical columns
# - scaler applied to numerical columns
preprocessor = ColumnTransformer([
    ('onehot', onehot, categorical_columns),
    ('zscore', scaler, numerical_columns)
])

# Build a full ML pipeline:
# - preprocessing: applies the ColumnTransformer to the data
# - classifier: K-Nearest Neighbors classifier trained on the processed data
diabetes_clf = Pipeline([
    ('preprocessing', preprocessor),
    ('classifier', KNeighborsClassifier())
])
````

### Perform Grid Search for Hyperparameter Tuning

Now that we’ve built our machine-learning pipeline, our next step is to tune the hyperparameters of our KNN model so we can find the version that performs best on our data.

We’ll use **`GridSearchCV`**, which tries out different combinations of hyperparameters and selects the one that gives the highest score.

`GridSearchCV` requires a **parameter grid** — a dictionary where:

- each **key** is the name of a parameter we want to tune  
- each **value** is a list of possible settings for that parameter  

Since we are using a pipeline, we refer to model parameters with this format: `step_name__parameter_name` In our pipeline, the KNN classifier is named `"classifier"`, so the parameter we want to tune becomes: `classifier__n_neighbors`.


For this demo, we will tune the number of neighbors the KNN model uses. We will test these values: **3, 5, 7, 10, 15**

Create a dictionary named **`param_grid`** with:

- key: `"classifier__n_neighbors"`
- value: the list of neighbor values above.

```python
param_grid = {
    "classifier__n_neighbors": [3, 5, 7, 10, 15]
}

````python
param_grid = {'classifier__n_neighbors' : [3, 5, 7, 10, 15]}
````

Initialize `GridSearchCV` with `diabetes_clf`, `param_grid`, `cv = 5`, and `verbose = 1`. Name the variable as `best_model`. Train `best_model` on `X_train`, `y_train`.

````python
best_model = GridSearchCV(diabetes_clf, param_grid, cv=5, verbose=1)
best_model.fit(X_train, y_train)
````

````python
diabetes_predicted_test_Y = best_model.predict(X_test)
class_report_test = classification_report(y_test,diabetes_predicted_test_Y)
print(class_report_test)s
````
