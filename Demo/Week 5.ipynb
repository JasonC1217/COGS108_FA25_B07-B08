## Part I: Data Wrangling

Here, we have a predefined function to standarize heights data:

````python
def standardize_height(str_in):
    """
    Convert a height string to centimeters.
    
    This function takes a height string as input, which can be in inches or centimeters,
    and returns the height standardized to centimeters. If the input is invalid or not
    recognized, it returns a NaN value.
    
    Parameters:
    - str_in (str): A string representing a height, e.g., "5 inches" or "12 cms".
    
    Returns:
    - float: The height in centimeters or NaN if input is invalid.
    """
    
    try:
        # Convert input string to lowercase to ensure consistent processing
        str_in = str_in.lower()
        
        # Remove any leading or trailing whitespaces
        str_in = str_in.strip()

        # Check if input string contains 'in' or 'inches' indicating height in inches
        if 'in' in str_in or 'inches' in str_in:
            # Remove the word 'inches' and 'in' from the string to isolate the number
            str_in = str_in.replace('inches', '')
            str_in = str_in.replace('in', '')
            
            # Convert the isolated number string to float and then convert inches to centimeters
            output = float(str_in)
            output = 2.54 * output
        # Check if input string contains 'cms' or 'centimeters' indicating height in centimeters
        elif 'cms' in str_in or 'centimeters' in str_in:
            # Remove the word 'cms' and 'centimeters' from the string to isolate the number
            str_in = str_in.replace('cms', '')
            str_in = str_in.replace('centimeters', '')
            
            # Convert the isolated number string to float
            output = float(str_in)
        else:
            # If neither 'inches' nor 'cms' is present, return NaN
            output = np.nan
    except:
        # If any error occurs during processing (e.g., converting a non-number string to float), return NaN
        output = np.nan
        
    return output
````

Now we have a function defined. We can just apply this function to all elements in a single column using .apply():

````python
df['height'] = df['height'].apply(standardize_height)
````

### Specifically, we will need:
#### Input
A string representing a weight, e.g.:
- `"150 pounds"`

#### Output
A `float` representing the weight in kilograms, or `np.nan` if the input is invalid.

#### Logic
1. The input string is converted to lowercase and stripped of extra spaces.

3. If none of the known units are found or conversion fails, the function returns `np.nan`.

#### Edge Case Handling

Uses a `try-except` block to catch formatting errors or invalid numbers.
Returns `np.nan` for unrecognized formats or conversion failures.

#### Examples
```python

````python
def standardize_weight(str_in):
    """
    Convert a weight string to kilograms.
    
    This function takes a weight string as input, which can be in pounds or kilograms,
    and returns the weight standardized to kilograms. If the input is invalid or not
    recognized, it returns a NaN value.
    
    Parameters:
    - str_in (str): A string representing a weight, e.g., "150 pounds" or "68 kgs".
    
    Returns:
    - float: The weight in kilograms or NaN if the input is invalid.
    """
    
    try:
        # Convert the input string to lowercase to ensure consistent processing
        str_in = str_in.lower()
        
        # Remove any leading or trailing whitespaces
        str_in = str_in.strip()

        # Check if the input string contains 'pounds' or 'lbs' indicating weight in pounds
        if 'pounds' in str_in or 'lbs' in str_in:
            # Remove the word 'pounds' and 'lbs' from the string to isolate the number
            str_in = str_in.replace('pounds', '')
            str_in = str_in.replace('lbs', '')
            
            # Convert the isolated number string to float and then convert pounds to kilograms
            output = float(str_in)
            output = 0.453592 * output
            
        # Check if the input string contains 'kgs' or 'kilograms' indicating weight in kilograms
        elif 'kgs' in str_in or 'kilograms' in str_in:
            # Remove the word 'kgs' and 'kilograms' from the string to isolate the number
            str_in = str_in.replace('kgs', '')
            str_in = str_in.replace('kilograms', '')
            
            # Convert the isolated number string to float
            output = float(str_in)
        else:
            # If neither 'pounds' nor 'kgs' is present, return NaN
            output = np.nan
    except:
        # If any error occurs during processing (e.g., converting a non-number string to float), return NaN
        output = np.nan
        
    return output
````

#### What is a t-test?

A **t-test** is a statistical hypothesis test used to determine whether there is a **significant difference between the means** of two groups. It is especially useful when dealing with **small sample sizes** and when the population standard deviation is **unknown**. The test assumes that the data is approximately **normally distributed**.

There are several types of t-tests, including:
- **Independent t-test (unpaired):** Compares means from two **independent** groups (e.g., control vs. treatment).
- **Paired t-test:** Compares means from the **same group at different times** (e.g., before vs. after treatment).
- **One-sample t-test:** Compares the mean of a **single group** to a known or hypothesized population mean.

#### T-Test Formula

For an **independent two-sample t-test**, the test statistic is calculated as:

#### $t = \frac{\bar{X}_1 - \bar{X}_2}{\sqrt{\frac{s_1^2}{n_1} + \frac{s_2^2}{n_2}}}$

Where:
- $ \bar{X}_1, \bar{X}_2 $: sample means of group 1 and group 2  
- $ s_1^2, s_2^2 $: sample variances of group 1 and group 2  
- $ n_1, n_2 $: sample sizes of group 1 and group 2  

This formula compares the difference in means relative to the **pooled standard error**.

#### Interpretation
- The **null hypothesis (H₀)** typically states that there is **no difference** between the group means:  
  $ H_0: \mu_1 = \mu_2
  $
- The **alternative hypothesis (H₁)** suggests that the means **are different**:  
  $
  H_1: \mu_1 \ne \mu_2
  $

A **large absolute value of the t-statistic** (along with a small p-value) indicates strong evidence **against** the null hypothesis.

#### When to Use a T-Test
- Comparing test scores of two classes
- Evaluating the effect of a treatment
- Testing marketing campaign outcomes across user groups

> Note: Always check assumptions (normality, equal variance, independence) before applying a t-test.

#### ANOVA: Analysis of Variance

When we have multiple **t-tests** looking at the same question, there is a higher risk of getting a **false positive** result than when we only run one test. This happens because each additional test increases the overall probability of finding a significant result purely by chance.

To handle this, we consider the **family-wise error rate (FWER)** — the probability that **at least one** of the $n$ tests incorrectly rejects the null hypothesis (i.e., at least one difference is due to chance alone).

A family of $n$ t-tests yields a family-wise error rate of:

$$
p_\text{family} = 1 - \prod_{i=1}^n (1 - p_i)
$$

This value grows quickly as the number of tests increases.  
For example, if we run 10 tests each at $p = 0.05$, the family-wise error rate becomes:

$$
p_\text{family} = 1 - (1 - 0.05)^{10} \approx 0.40
$$

That means there’s a **40% chance** of finding a false positive somewhere in the set of tests!

---

#### Multiple Comparison Corrections

There are statistical methods designed to correct for this issue, known as **multiple comparison corrections**, such as:

- **Bonferroni correction**
- **Benjamini–Hochberg procedure**

These approaches adjust p-values (or significance thresholds) so that the reported results represent the **family-wise error rate**, reducing the risk of false discoveries.

---

#### Why Use ANOVA?

An even better solution, when comparing **more than two groups**, is to use a **test that directly accounts for the family-wise error rate** — such as the **one-way ANOVA (Analysis of Variance)**.

Instead of asking:

> “For all pairs of countries, is there any difference between each pair?”  
> (which would require multiple t-tests)

We ask:

> “Is there any difference between the countries as a whole?”

ANOVA won’t tell us *which* countries differ from each other, but it does tell us *whether there are any significant differences at all*.

---

#### Concept and Intuition

A one-way ANOVA compares:
- **Between-group variance:** How much the group means differ from the overall mean.
- **Within-group variance:** How much the individual observations vary within each group.

![ANOVA_example](ANOVA.png) 

If most of the total variance is **explained by group differences**, then group membership (e.g., nationality) matters.

The ANOVA test computes an **F-statistic**:

$$
F = \frac{\text{Mean Between-group variance}}{\text{Mean Within-group variance}}
$$

A larger $F$ value indicates that the group means are more spread out relative to the variability within each group. This $F$-statistic is then converted into a **p-value**, similar to how the *t*-statistic is used in a t-test.

---

#### Performing ANOVA in Python

We can use the `statsmodels` library to perform a one-way ANOVA.  

This will output a table showing the **F-statistic**, **degrees of freedom**, **and p-value**, allowing us to test whether nationality has a significant effect on players’ heights.

````python
import statsmodels.api as sm

nats = ['Japan','Spain','England','France','Germany','Netherlands']
df_nats = df.query("nationality in @nats")

anova_result = sm.stats.anova_oneway(data=df_nats['height'],groups=df_nats['nationality'])

print(anova_result)
````

#### Post-hoc Analysis: Identifying Which Groups Differ

If the ANOVA test shows a **significant effect**, it tells us that *at least one group mean* differs from the others — but **it doesn’t tell us which groups** are different.

To identify **specific pairwise differences**, we use **post-hoc tests**.  
These tests perform multiple pairwise comparisons while controlling for the **family-wise error rate**, preventing false positives due to multiple testing.

Common post-hoc methods include:
- **Tukey’s Honest Significant Difference (HSD)** test  
- **Bonferroni (or Benjamini-Hochberg)-adjusted pairwise t-tests**

Among these, **Tukey’s HSD** is widely used for one-way ANOVA because it adjusts p-values automatically and provides confidence intervals for group mean differences.

---

#### Performing Tukey’s HSD Test in Python

We can use the `pairwise_tukeyhsd` function from `statsmodels` to perform this analysis.

````python
from statsmodels.stats.multicomp import pairwise_tukeyhsd

tukey_result = pairwise_tukeyhsd(
    endog = df_nats['height'],       # Dependent variable (the numeric column we're comparing)
    groups = df_nats['nationality'], # Categorical variable defining the groups
    alpha = .05                     # Significance level (5%)
)

# Display the summary table of results
# ------------------------------------
# The output includes:
# - The pairs of nationalities being compared
# - The mean difference between their heights
# - The adjusted p-value accounting for multiple tests
# - Confidence intervals for the difference
# - A "reject" column (True if the difference is statistically significant)
````
